### Сравнение табличных и расчетных значений rougher.output.recovery

По формуле рассчитаем эффективность обогащения золота и сравним рассчетные данные с табличными методом средней абсолютной ошибки **МАЕ**.

<img src='https://pictures.s3.yandex.net/resources/Recovery_1576238822.jpg'>

где:

    C — доля золота в концентрате после флотации/очистки;
    F — доля золота в сырье/концентрате до флотации/очистки;
    T — доля золота в отвальных хвостах после флотации/очистки.
<img src='https://pictures.s3.yandex.net/resources/MAE2-84f43ddf-9489-420a-9064-672c99fccffe_1574260729.jpg'>

```python
C = gold_recovery_train['rougher.output.concentrate_au']
F = gold_recovery_train['rougher.input.feed_au']
T = gold_recovery_train['rougher.output.tail_au']
```

```python
recovery = C * (F - T) / (F * (C - T)) * 100
```

Для обеспечения возможности сравнения удалим пропуски из признака **rougher.output.recovery** и передадим полученный объект в переменную `rougher_output_recovery_train`.

```python
rougher_output_recovery_train = gold_recovery_train['rougher.output.recovery'].dropna()
```

```python
recovery = pd.Series(recovery, index = rougher_output_recovery_train.index)
```

Вычислим среднюю абсолютную ошибку.

```python
mae(recovery, rougher_output_recovery_train)
```

```python
recovery.mean()
```

Получили среднюю абсолютную ошибку **(MAE)**, равную `9.45 * 10^(-15)`. Среднее значение показателя, измеряемое в величинах десятичного порядка, говорит о том, что средняя абсолютная ошибка ничтожно мала, можем говорить о **совпадении табличных данных результатам расчета по формуле.**

### Предобработка данных

```python
#сохраним в отдельную переменную выборку gold_recovery_train до обработки
gold_recovery_train_original = gold_recovery_train.copy()
```

Сперва избавимся от столбцов выборки `_train`, которые отсутствуют в выборке `_test` (приведем количество признаков к равенству).

```python
gold_recovery_train = gold_recovery_train.drop(delta_columns, axis=1)
```

Также нам необходимо добавить к выборкам `_train` и `_test` целевые признаки из совокупности данных `_full`.

```python
gold_recovery_train = gold_recovery_train.merge(
    gold_recovery_full[['final.output.recovery', 'rougher.output.recovery', 'date']], on='date')
gold_recovery_test = gold_recovery_test.merge(
    gold_recovery_full[['final.output.recovery', 'rougher.output.recovery', 'date']], on='date')
```

Поскольку данные индексируются датой и временем получения информации (признак `date`), и по условию техпроцесса соседние данные часто являются схожими, для всех трех таблиц заполним пропуски предыдущим значением с помощью метода `.ffill(axis=0)`, предварительно избавившись от строк с пропусками в целевых значениях.

```python
print('До удаления пропусков:', gold_recovery_train.shape[0], 'записей.')
gold_recovery_full = gold_recovery_full.dropna(subset=['final.output.recovery', 'rougher.output.recovery'])
gold_recovery_test = gold_recovery_test.dropna(subset=['final.output.recovery', 'rougher.output.recovery'])
gold_recovery_train = gold_recovery_train.dropna(subset=['final.output.recovery', 'rougher.output.recovery'])
print('После удаления пропусков:', gold_recovery_train.shape[0], 'записей.')

gold_recovery_test = gold_recovery_test.ffill(axis=0)
gold_recovery_train = gold_recovery_train.ffill(axis=0)
gold_recovery_full = gold_recovery_full.ffill(axis=0)
```

```python
print('Количество пропусков в данных после заполнения предыдущим значением:',
     gold_recovery_test.isna().sum().sum() + gold_recovery_train.isna().sum().sum() +
     gold_recovery_full.isna().sum().sum())

print('Размерность таблиц составляет:', gold_recovery_full.shape, gold_recovery_train.shape,
     gold_recovery_test.shape)
```

Таким образом, мы избавились от пропусков в значениях признаков по объектам в наших исходных данных, а также заполнили отсутствующие признаки выборки `_test` заимствованием из выборки `_full`.

### Функция sMAPE

Напишем функцию `smape` для вычисления симметричной средней абсолютной процентной ошибки по формуле. Она похожа на **MAE**, но выражается не в абсолютных величинах, а в относительных, а также одинаково учитывает масштаб и целевого признака, и предсказания.

<img src='https://pictures.s3.yandex.net/resources/smape_1576239058.jpg'>

Итоговое значение sMAPE на 25% состоит из показателя для этапа флотации, и на 75% из показателя финального этапа.
<img src='https://pictures.s3.yandex.net/resources/_smape_1576239054.jpg'>

```python
def smape(y_true, y_pred):
    return 1 / len(y_true) * np.sum(np.abs(y_true - y_pred) / 
                  ((np.abs(y_true) + np.abs(y_pred)) / 2)) *100
```

```python
#видоизмененная функция для передачи в make_scorer
def smape_score(y_true, y_pred):
    smape_0 = 1 / len(y_true) * np.sum(np.abs(y_true - y_pred) / 
                  ((np.abs(y_true) + np.abs(y_pred)) / 2)) *100
    return smape_0[0] *.75 + smape_0[1] * .25
```

### Выбор модели на основе кросс-валидации

Поскольку целевым признаком является непрерывная величина, мы имеем дело с задачей регрессии. То есть нам подходят такие модели, как линейная регрессия, дерево регрессии и случайный лес регрессии.

Для оценки результата кросс-валидации передадим в параметр `scoring` оценку по **sMAPE**.

```python
smape_score = make_scorer(smape_score,  greater_is_better=False)
```

#### Случайный лес регрессии

*// Время выполнения блока кода ниже - 20 минут.

```python
%%time 
best_score = -50
best_nest = 0
for n_est in range(10, 100, 10):
    forest_model = RandomForestRegressor(n_estimators=n_est, random_state=rs)
    forest_model_score = cross_val_score(forest_model, features_train, target_train,
                                     cv=5, scoring=smape_score)
    if forest_model_score.mean() > best_score:
        best_score = forest_model_score.mean()
        best_nest = n_est

print('Наиболее оптимальное значение параметра n_estimators:', best_nest)
print('Список значение sMAPE по блокам кросс-валидации:', forest_model_score)
print('Итоговое значение sMAPE:', forest_model_score.mean())
```

При выполнении кросс-валидации установили лучшее значение параметра `n_estimator = 90`.

```python
%%time
l_model = LinearRegression()
l_model.fit(features_train, target_train)
l_predictions = l_model.predict(features_test)
l_predictions_final = pd.Series([])
l_predictions_rougher = pd.Series([])

for i in range(l_predictions.shape[0]):
    l_predictions_final[i] = l_predictions[i][0]
    l_predictions_rougher[i] = l_predictions[i][1]

l_result_rougher = mae(target_test['rougher.output.recovery'], l_predictions_rougher)
l_result_final = mae(target_test['final.output.recovery'], l_predictions_final)
l_sMAPE_final = smape(target_test['rougher.output.recovery'], 
                      l_predictions_rougher) * .25 + smape(target_test['final.output.recovery'], 
                                                           l_predictions_final) * .75
    
print('MAE по этапу флотации:', l_result_rougher)
print('МАЕ на финальном этапе:', l_result_final)
print('Значение sMAPE:', l_sMAPE_final)
```
